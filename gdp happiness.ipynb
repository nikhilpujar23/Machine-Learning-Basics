```python
h_index=pd.read_csv(r'D:\4th sem\machine learning/BLI.csv',thousands=',')
```

```python
gdp.rename(columns={'2015':'Gdp_per_capita'},inplace=True)
```

```python
gdp.set_index('Country Name',inplace=True)
```

```python
gdp.index
```

```python
h_index=h_index[h_index['INEQUALITY']=='TOT']
h_index
```

```python

h_index = h_index.pivot(index="Country" , columns="Indicator", values="Value")
```

```python
h_index
```

```python
df=pd.merge(
left=h_index,right=gdp,left_index=True,right_index=True
)
df=df[["Gdp_per_capita", 'Life satisfaction']]
df
```

```python
import matplotlib.pyplot as plt

X=np.c_[df['Gdp_per_capita']]
Y=np.c_[df['Life satisfaction']]
df.plot(kind='scatter',x='Gdp_per_capita',y='Life satisfaction')
plt.show()

#np.array won't work
```

```python
print(X,Y)
```

```python
import sklearn.linear_model
model = sklearn.linear_model.LinearRegression()
# Train the model
model.fit(X, Y)
# Make a prediction for Cyprus
X_new = [[22587]] 
print(model.predict(X_new))
```

```python
import sklearn.neighbors
model=sklearn.neighbors.KNeighborsRegressor(n_neighbors=3)
x_new=[[22587]]
model.fit(X,Y)
print(model.predict(x_new))
```
